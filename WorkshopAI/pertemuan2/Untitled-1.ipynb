```python
from skimage.io import imread, imshow
#import matplotlib.pyplot as plt
#import numpy as np
import cv2 
src = "gambar.jpg"
```

```python
#image
imageN = imread(src)
imshow(imageN)
```

```python
#image gray
imageGray = imread(src, as_gray=True)
imshow(imageGray)
```

```python
#image color
imageN2 = imread(src)
imageConvert = cv2.cvtColor(imageN2, cv2.COLOR_BGR2RGB)
imshow(imageConvert)
cv2.imwrite("test_write.jpg", imageN2)
```

```python
#image crop
imageN3 = imread(src)
imageCrop = imageN3[420:630, 420:710]
imshow(imageCrop)
imageCrop = cv2.cvtColor(imageCrop, cv2.COLOR_RGB2BGR)
cv2.imwrite("image_crop.jpg", imageCrop)
```

```python
#resize image
imageN4 = imread(src)
width, height = imageN4.shape[ :2]

img1 = cv2.resize(imageN4, (width + 50, height - 20), interpolation=cv2.INTER_CUBIC)
img2 = cv2.resize(imageN4, (width // 2, height // 2), interpolation=cv2.INTER_AREA)
img3 = cv2.resize(imageN4, (width * 2, height))

cv2.imshow("INTER CUBIC", img1)
cv2.waitKey(0)
cv2.imshow("INTER AREA", img2)
cv2.waitKey(0)
cv2.imshow("INTER LINEAR", img3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
